**1. What is the minimum, median, max, and average salary for employee by Married Description and Gender.**

  Hints: use agg function

```python
# 1. What is the minimum, median, max, and average salary for employee by Married Description and Gender
# Hints: use agg function
# Expected Output as below
# (5 Points)
```

**Jawab**

```python
# Group by MaritalDesc and Sex, and then calculate the required aggregations
result = data.groupby(['MaritalDesc', 'Sex'])['Salary'].agg(['min', 'median', 'max', 'mean'])

# Rename the columns for clarity
result = result.rename(columns={'min': 'min_salary', 'median': 'median_salary', 'max': 'max_salary', 'mean': 'average_salary'})

# Reset the index to match the expected output format
result.reset_index(inplace=True)

# Display the result
result
```

> - average_salary: Gaji rata-rata (mean) dalam setiap kelompok.

```python
# Filter out rows where TermReason is not 'N/A-StillEmployed' (i.e., terminated employees)
terminated_data = data[data['TermReason'] != 'N/A-StillEmployed']

# Use value_counts to count the occurrences of each termination reason
termination_counts = terminated_data['TermReason'].value_counts()

# Select the top 5 reasons
top_5_termination_reasons = termination_counts.head(5)

# Rename the Series to 'total_termination' and reset the index
top_5_termination_reasons = top_5_termination_reasons.rename('total_termination').reset_index()

# Rename the 'index' column to 'TermReason'
top_5_termination_reasons = top_5_termination_reasons.rename(columns={'index': 'TermReason'})

# Display the top 5 termination reasons
top_5_termination_reasons
```

**Kesimpulan :**

Kode tersebut dirancang untuk melakukan analisis data terkait alasan pengakhiran pekerjaan (termination reasons).

1. terminated_data = data[data['TermReason'] != 'N/A-StillEmployed']: Langkah ini menyaring (filter) baris data yang memiliki nilai kolom 'TermReason' yang bukan 'N/A-StillEmployed'. Dengan kata lain, hanya data pekerja yang telah diakhiri (terminated) yang akan dipertimbangkan dalam analisis selanjutnya.

2. termination_counts = terminated_data['TermReason'].value_counts(): Setelah penyaringan, kode ini menghitung berapa kali setiap alasan pengakhiran muncul dalam data yang telah disaring. Hasilnya adalah objek Series yang berisi jumlah munculnya setiap alasan pengakhiran.

3. top_5_termination_reasons = termination_counts.head(5): Langkah ini memilih lima alasan pengakhiran teratas dari data yang telah dihitung sebelumnya.

4. top_5_termination_reasons = top_5_termination_reasons.rename('total_termination').reset_index(): Kode ini mengubah nama objek Series dari 'top_5_termination_reasons' menjadi 'total_termination' dan mengatur ulang indeksnya. Tujuannya adalah agar data ini memiliki format yang sesuai dengan output yang diinginkan.

5. top_5_termination_reasons = top_5_termination_reasons.rename(columns={'index': 'TermReason'}): Langkah ini digunakan untuk mengubah nama kolom 'index' menjadi 'TermReason' agar sesuai dengan nama kolom yang diharapkan pada output.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grouping the data by 'Department' and counting the number of terminations for each department
termination_counts = data['Department'].value_counts().sort_values(ascending=True)

# Creating a different color palette for each bar
colors = sns.color_palette('inferno', len(termination_counts))

# Creating a horizontal bar chart
plt.figure(figsize=(10, 6))
ax = termination_counts.plot(kind='barh', color=colors)

# Adding labels on the top-right of each bar
for i, count in enumerate(termination_counts):
    plt.text(count + 1, i, str(count), va='center', fontsize=10, fontweight='bold', color='black')

# Adding labels and a title
plt.xlabel('Termination Count')
plt.ylabel('Department')
plt.title('Employee Terminations by Department')

# Creating a legend
legend_labels = termination_counts.index  # Taking department names as legend labels
legend = plt.legend(legend_labels, title='Departments', loc='lower right')

# Displaying the horizontal bar chart with the legend
plt.show()
```

Kesimpulan :

1. Data dikelompokkan berdasarkan kolom 'Department' dan dihitung jumlah pengakhiran untuk setiap departemen menggunakan data['Department'].value_counts().

2. Sebuah palet warna yang berbeda, didasarkan pada palet 'husl' dari Seaborn, .dibuat dengan panjang yang sesuai dengan jumlah departemen yang ada.

3. Visualisasi horizontal bar chart dibuat dengan plt.barh() dengan menggunakan data penghitungan pengakhiran dan palet warna yang telah dibuat sebelumnya.

4. Loop for digunakan untuk menambahkan label di kanan atas setiap bar yang menampilkan jumlah pengakhiran.

7. Analisis Korelasi: Hitung koefisien korelasi untuk mengukur kekuatan dan arah hubungan antara ID manajer dan skor kinerja.

Berikut adalah contoh kode Python menggunakan Pandas dan Seaborn untuk melakukan beberapa langkah ini:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
pd.set_option('display.max_columns', None)

# Load your dataset into a Pandas DataFrame
data = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/HRDataset_v14.csv')
```

```python
# Group data by ManagerID and calculate mean performance scores
manager_performance = data.groupby('ManagerID')['PerformanceScore'].mean().reset_index()

# Create a bar chart to visualize the mean performance scores
plt.figure(figsize=(10, 6))
sns.barplot(x='ManagerID', y='PerformanceScore', data=manager_performance)
plt.xlabel('ManagerID')
plt.ylabel('Mean Performance Score')
plt.title('Mean Performance Score by Manager')
plt.xticks(rotation=45)
plt.show()

# Perform ANOVA to test for significant differences in performance scores
f_statistic, p_value = stats.f_oneway(*[group['PerformanceScore'] for name, group in data.groupby('ManagerID')])
print(f'F-Statistic: {f_statistic}')
print(f'P-Value: {p_value}')

# You can also calculate correlation coefficients
correlation_coefficient = data['ManagerID'].corr(data['PerformanceScore'])
print(f'Correlation Coefficient: {correlation_coefficient}')
```

Kode ini memberikan contoh dasar tentang cara menjelajahi dan menganalisis hubungan antara manajer dan skor kinerja menggunakan Python.

## **20. What are our best recruiting sources if we want to ensure a low ratio of termination?**

Untuk menentukan sumber rekrutmen terbaik yang dapat memastikan rasio pengakhiran yang rendah, dapat mengikuti langkah-langkah berikut menggunakan Python:

1. Filter Data: Mulailah dengan menyaring dataset untuk hanya mencakup karyawan yang telah diakhiri (terminated). Ini akan memungkinkan menganalisis rasio pengakhiran berdasarkan sumber rekrutmen di antara mereka yang telah diakhiri.

2. Hitung Pengakhiran berdasarkan Sumber: Gunakan dataset yang telah disaring untuk menghitung jumlah pengakhiran untuk setiap sumber rekrutmen.

3. Hitung Rasio: Hitung rasio pengakhiran untuk setiap sumber rekrutmen dengan membagi jumlah pengakhiran dengan total jumlah penerimaan dari sumber tersebut.

4. Urutkan dan Pilih: Urutkan sumber rekrutmen berdasarkan rasio pengakhiran mereka secara berurutan dari yang terkecil. Sumber-sumber dengan rasio pengakhiran terendah akan menjadi sumber rekrutmen terbaik Anda untuk memastikan rasio pengakhiran yang rendah.

Berikut contoh kode Python:

```python
import pandas as pd

# Filter data to include only terminated employees
terminated_data = data[data['Termd'] == 1]

# Count terminations by recruitment source
termination_counts = terminated_data['RecruitmentSource'].value_counts()

# Calculate termination ratios
total_hires_by_source = data['RecruitmentSource'].value_counts()
termination_ratios = termination_counts / total_hires_by_source

# Sort recruitment sources by termination ratio in ascending order
sorted_recruitment_sources = termination_ratios.sort_values(ascending=True)

# Display the best recruiting sources with the lowest termination ratios
print(sorted_recruitment_sources)
```

Kode ini akan memberikan daftar sumber rekrutmen yang diurutkan berdasarkan rasio pengakhiran mereka secara berurutan dari yang terendah. Sumber-sumber di bagian atas daftar akan menjadi sumber rekrutmen terbaik Anda untuk memastikan rasio pengakhiran yang rendah.
